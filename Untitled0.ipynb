```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.preprocessing import LabelEncoder
import networkx as nx
import torch
import torch.nn as nn
import torch.nn.functional as F

# -----------------------------
# LOAD DATA
# ---------------------C:\\Users\\10\\OneDrive\\Documents\\Fraud_detection\\User0_credit_card_transactions.csv--------
df = pd.read_csv("/User0_credit_card_transactions.csv.zip")
FAST_MODE = True

if FAST_MODE:
    df = df.sample(3000, random_state=42)

df.info()

# -----------------------------
# FRAUD COUNT
# -----------------------------
print(df[df['Is Fraud?'] == 'Yes'].shape)

# -----------------------------
# MISSING VALUE VISUALIZATION (1)
# -----------------------------
percent_missing = (df.isnull().sum() * 100 / df.shape[0]).sort_values(ascending=False)

plt.figure(figsize=(10,5))
plt.bar(percent_missing.index, percent_missing.values, edgecolor='black')
plt.title("Missing Value Analysis")
plt.ylabel("% Missing")
plt.xticks(rotation=90)
plt.show()

# -----------------------------
# FEATURE ENGINEERING
# -----------------------------
df["card_id"] = df["User"].astype(str) + "_" + df["Card"].astype(str)
df["Amount"] = df["Amount"].str.replace("$","").astype(float)
df["Hour"] = df["Time"].str[0:2].astype(int)
df["Minute"] = df["Time"].str[3:5].astype(int)

df = df.drop(["Time","User","Card"], axis=1)
df["Errors?"] = df["Errors?"].fillna("No error")
df = df.drop(columns=["Merchant State","Zip"])

df["Is Fraud?"] = df["Is Fraud?"].apply(lambda x: 1 if x == "Yes" else 0)

df["Merchant City"] = LabelEncoder().fit_transform(df["Merchant City"])
df["Use Chip"] = LabelEncoder().fit_transform(df["Use Chip"])
df["Errors?"] = LabelEncoder().fit_transform(df["Errors?"])

# -----------------------------
# FRAUD DISTRIBUTION (2)
# -----------------------------
plt.figure(figsize=(5,4))
df["Is Fraud?"].value_counts().plot(kind='bar', color=['green','red'])
plt.title("Fraud vs Non-Fraud Distribution")
plt.xticks([0,1], ["Non-Fraud","Fraud"], rotation=0)
plt.show()

# -----------------------------
# AMOUNT DISTRIBUTION (3)
# -----------------------------
plt.figure(figsize=(7,4))
plt.hist(df["Amount"], bins=50)
plt.yscale("log")
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency (log)")
plt.show()

# -----------------------------
# FRAUD vs AMOUNT (4)
# -----------------------------
plt.figure(figsize=(7,4))
plt.hist(df[df["Is Fraud?"]==0]["Amount"], bins=50, alpha=0.6, label="Non-Fraud")
plt.hist(df[df["Is Fraud?"]==1]["Amount"], bins=50, alpha=0.6, label="Fraud")
plt.legend()
plt.yscale("log")
plt.title("Fraud vs Non-Fraud Amounts")
plt.show()

# -----------------------------
# FRAUD BY HOUR (5)
# -----------------------------
fraud_by_hour = df.groupby("Hour")["Is Fraud?"].mean()
plt.figure(figsize=(8,4))
plt.plot(fraud_by_hour.index, fraud_by_hour.values, marker='o')
plt.title("Fraud Probability by Hour")
plt.xlabel("Hour")
plt.ylabel("Fraud Rate")
plt.grid(True)
plt.show()

# -----------------------------
# GRAPH CONSTRUCTION
# -----------------------------
G = nx.MultiGraph()

G.add_nodes_from(df["card_id"].unique(), type="card")
G.add_nodes_from(df["Merchant Name"].unique(), type="merchant")

for _, row in df.iterrows():
    G.add_edge(
        row["card_id"],
        row["Merchant Name"],
        year=row["Year"],
        month=row["Month"],
        day=row["Day"],
        hour=row["Hour"],
        minute=row["Minute"],
        amount=row["Amount"],
        use_chip=row["Use Chip"],
        merchant_city=row["Merchant City"],
        errors=row["Errors?"],
        mcc=row["MCC"]
    )

print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())

# -----------------------------
# DEGREE DISTRIBUTION (6)
# -----------------------------
degrees = [d for _, d in G.degree()]
plt.figure(figsize=(7,4))
plt.hist(degrees, bins=50)
plt.yscale("log")
plt.title("Graph Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Count (log)")
plt.show()

# -----------------------------
# GRAPH VISUALIZATION (7)
# -----------------------------
plt.figure(figsize=(8,8))
subG = G.subgraph(list(G.nodes())[:50])
pos = nx.spring_layout(subG, seed=42)
nx.draw(subG, pos, node_size=50, alpha=0.7)
plt.title("Sample Transaction Graph")
plt.show()

# -----------------------------
# EDGE PROPERTY DISTRIBUTION (8)
# -----------------------------
edge_errors = nx.get_edge_attributes(G, "errors")
plt.figure(figsize=(6,4))
plt.bar(*np.unique(list(edge_errors.values()), return_counts=True))
plt.title("Edge Error Distribution")
plt.xlabel("Error Code")
plt.ylabel("Count")
plt.show()

# -----------------------------
# PREPARE DATA FOR MODEL
# -----------------------------
edge_list = list(G.edges(data=True))

x = []
for edge in edge_list:
    vals = list(edge[2].values())
    vals = [float(v) for v in vals]
    x.append(vals)

x = torch.tensor(x, dtype=torch.float)
y = torch.tensor(df["Is Fraud?"].values[:len(x)], dtype=torch.float)

# -----------------------------
# MODEL
# -----------------------------
class FraudGNN(nn.Module):
    def __init__(self, in_dim, hidden):
        super().__init__()
        self.fc1 = nn.Linear(in_dim, hidden)
        self.fc2 = nn.Linear(hidden, 1)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        return self.fc2(x).squeeze()

model = FraudGNN(x.shape[1], 16)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.BCEWithLogitsLoss()

# -----------------------------
# TRAINING (9)
# -----------------------------
loss_history = []

for epoch in range(200):
    optimizer.zero_grad()
    out = model(x)
    loss = criterion(out, y)
    loss.backward()
    optimizer.step()
    loss_history.append(loss.item())

    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss {loss.item():.4f}")

plt.figure(figsize=(7,4))
plt.plot(loss_history)
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# -----------------------------
# PREDICTION VISUALIZATIONS (10)
# -----------------------------
with torch.no_grad():
    preds = torch.sigmoid(model(x)).numpy()

plt.figure(figsize=(7,4))
plt.hist(preds, bins=50)
plt.title("Predicted Fraud Probability Distribution")
plt.xlabel("Probability")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(preds, y.numpy(), alpha=0.3)
plt.title("Prediction vs Actual")
plt.xlabel("Predicted Probability")
plt.ylabel("Actual Fraud")
plt.show()
```
